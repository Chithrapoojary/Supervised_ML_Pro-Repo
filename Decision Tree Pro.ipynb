```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("diabetes.csv")   #Reading  the dataset
```

```python
data.head() #checking the first 5 rows of the dataset
```

```python
data.isna().sum()  #checking null values and categorical values
```

```python
x=data.drop(['Outcome'],axis=1) #dividing the data into dependent and independent variable
y=data['Outcome']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split     #splitting data into train and test test
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.85)
```

```python
from sklearn.tree import DecisionTreeClassifier  # Using Decision Tree Algo
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(x_train,y_train)   #training the model
```

```python
y_hat=classifier.predict(x_test)    #prediction using the model
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_hat)
```

```python
print(ac)    
```

```python
from sklearn.linear_model import LogisticRegression     #Using Logistic Regression algo
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_hat=model.predict(x_test)     #prediction using model
```

```python
acc=accuracy_score(y_test,y_hat)
```

```python
print(acc)
```
